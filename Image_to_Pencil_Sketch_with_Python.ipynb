### Import Required libraries
- Here I am using CV2 and matplotlib

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow  #--- to show image in colab
```

### Read and Display the Image

```python
org=cv2.imread("/content/drive/MyDrive/Colab Notebooks/AI & Deep Learning/Datasets/CNN/dog.jpg")
cv2_imshow(org)
```

### Convert above Image to Pencil Sketch

```python
#Covert to grey scale image
grey=cv2.cvtColor(org, cv2.COLOR_BGR2GRAY)

#Invert the image
invert=cv2.bitwise_not(grey)

#Blur the Image and Invert
blur=cv2.GaussianBlur(invert, (111,111),0)
invert_blur=cv2.bitwise_not(blur)

#Create Sketch using dividing grey image by inverted blured image
sketch=cv2.divide(grey, invert_blur, scale=256.0)

#Save Sketch
cv2.imwrite('sketch.png', sketch)

#Display Sketch
cv2_imshow(sketch)
```
